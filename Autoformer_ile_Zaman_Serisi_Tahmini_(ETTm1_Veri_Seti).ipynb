### Hücre 4: Zaman Serisi Veri Ön İşleme (Autoformer için)

* **Amaç:**
    ETTm1 veri setini Autoformer veya benzeri bir transformatör modelinin eğitimi için uygun formata getirmek. Bu, özellik seçimi, veri ölçeklendirme, veri setinin kronolojik olarak eğitim, doğrulama ve test alt kümelerine ayrılması ve modelin girdi olarak alacağı zaman dizisi pencerelerinin oluşturulmasını içerir.

* **Yapılan İşlemler:**
    1.  **Özellik Seçimi:**
        * Modelde kullanılacak özellikler seçilir. Bu çalışmada, hedef değişken olan 'OT' (Oil Temperature) ile birlikte diğer tüm mevcut sütunlar (HUFL, HULL, MUFL, MULL, LUFL, LULL) kovaryet (ek özellik) olarak kullanılacaktır. Eğer istenirse, özellik seçimi teknikleriyle daha az sayıda kovaryet de seçilebilir.
    2.  **Veri Bölme (Kronolojik):**
        * Zaman serisi verilerinde, gelecekteki değerleri tahmin etmeye çalıştığımız için veri sızıntısını (data leakage) önlemek amacıyla bölme işlemi kesinlikle kronolojik olarak yapılmalıdır.
        * Veri seti genellikle %60-%70 eğitim, %10-%20 doğrulama ve %10-%20 test şeklinde bölünür. Bu çalışmada, örneğin ilk %70'i eğitim, sonraki %10'u doğrulama ve son %20'si test seti olarak ayrılabilir. (Autoformer makalesindeki standart bölünme oranları 6:2:2 veya 7:1:2 gibidir.)
    3.  **Veri Ölçeklendirme (Normalization/Standardization):**
        * Derin öğrenme modelleri, genellikle girdilerin belirli bir aralıkta (örn: 0-1 veya ortalaması 0, standart sapması 1) olmasını tercih eder. Bu, eğitimin daha stabil ve hızlı olmasına yardımcı olur.
        * `sklearn.preprocessing.StandardScaler` (veriyi ortalaması 0, standart sapması 1 olacak şekilde ölçekler) veya `MinMaxScaler` (veriyi 0-1 arasına sıkıştırır) kullanılabilir.
        * **Önemli Not:** Ölçekleyici (`scaler`) **sadece eğitim verisi (`train_df`) üzerinde `fit_transform`** ile eğitilmeli ve öğrenilen bu ölçekleme parametreleri (ortalama ve standart sapma) daha sonra hem eğitim verisine hem de **doğrulama (`val_df`) ve test (`test_df`) verilerine sadece `transform`** metoduyla uygulanmalıdır. Bu, doğrulama ve test setlerinden eğitim setine bilgi sızmasını engeller.
    4.  **Zaman Dizisi Pencereleri Oluşturma (Windowing):**
        * Transformatör modelleri, sabit uzunlukta girdi dizileri (geçmiş gözlemler) ve tahmin edilecek sabit uzunlukta çıktı dizileri (gelecek gözlemler) ile çalışır.
        * Bir fonksiyon aracılığıyla, ölçeklendirilmiş eğitim, doğrulama ve test verilerinden girdi dizileri (`X`) ve karşılık gelen hedef diziler (`y`) oluşturulur.
        * Parametreler:
            * `sequence_length` (veya `seq_len`, `history_len`): Modele girdi olarak verilecek geçmiş zaman adımı sayısı (örn: 96 adım = 1 gün, 15 dakikalık veri için).
            * `prediction_length` (veya `pred_len`, `forecast_horizon`): Modelin tahmin edeceği gelecek zaman adımı sayısı (örn: ETTm1 için 96, 192, 336, 720 adımlar popülerdir).
        * Bu işlem, kayan bir pencere (sliding window) mantığıyla yapılır.

* **Uygulama Detayları/Sonuçlar (Beklenen):**
    * Veri setinin, belirtilen oranlarda kronolojik olarak `train_df`, `val_df` ve `test_df` olmak üzere üç DataFrame'e bölünmesi.
    * `StandardScaler`'ın sadece `train_df` üzerinde eğitilip, tüm DataFrame'lere uygulanması sonucu ölçeklendirilmiş verilerin (`scaled_train_data`, `scaled_val_data`, `scaled_test_data`) elde edilmesi.
    * Belirlenen `sequence_length` ve `prediction_length` kullanılarak, her bir alt küme için modelin girdisi olacak `X_train, X_val, X_test` (geçmiş diziler) ve hedefleri olacak `y_train, y_val, y_test` (gelecek diziler) NumPy dizilerinin oluşturulması.
    * Bu hücrenin sonunda, `X_train, y_train, X_val, y_val, X_test, y_test` ve ölçekleyici (`scaler`) objesi, model eğitimi için hazır olacaktır. Ayrıca, orijinal (ölçeklenmemiş) test verisi ve tahmin ufkundaki gerçek değerler, tahminlerin daha sonra geri ölçeklendirilmesi ve karşılaştırılması için saklanabilir.

```python
# Hücre 4: Zaman Serisi Veri Ön İşleme (Autoformer için)

from sklearn.preprocessing import StandardScaler

# df (yüklenmiş ve zaman indeksi ayarlanmış DataFrame) ve target_column ('OT')
# önceki hücreden gelmelidir.

if 'df' in globals() and df is not None and 'target_column' in globals() and target_column in df.columns:
    print("Veri ön işleme adımları başlatılıyor...")

    # 1. Özellik Seçimi (Tüm sütunları kullanalım)
    # Hedef değişkenimiz 'OT'. Diğerleri kovaryet.
    # Autoformer gibi modeller genellikle çok değişkenli (multi-variate) zaman serilerini işleyebilir.
    # Sütun sırası önemli olabilir. Genellikle hedef değişken sonda veya başta tutulur.
    # Şimdilik tüm sütunları kullanalım. Veri setindeki sütun sırası: HUFL,HULL,MUFL,MULL,LUFL,LULL,OT
    # Hedef sütun 'OT' zaten sonda.

    data_to_process = df.copy() # Orijinal df'i koruyalım
    feature_columns = data_to_process.columns.tolist()
    print(f"Kullanılacak özellikler (hedef dahil): {feature_columns}")

    # 2. Veri Bölme (Kronolojik)
    # Veri setinin boyutları: ~70000 kayıt
    # Örnek bölme oranları: %70 train, %10 validation, %20 test
    # Autoformer makalesindeki standart ETT bölünmesi:
    # ETTh1, ETTh2 için: 12 ay train, 4 ay val, 4 ay test (son 8 ay val+test)
    # ETTm1, ETTm2 için: İlk %60 train, sonraki %20 val, son %20 test (oranlar 6:2:2)

    num_total_samples = len(data_to_process)
    train_split_idx = int(num_total_samples * 0.6)
    val_split_idx = int(num_total_samples * 0.8) # %60 train + %20 val = %80

    train_df = data_to_process.iloc[:train_split_idx]
    val_df = data_to_process.iloc[train_split_idx:val_split_idx]
    test_df = data_to_process.iloc[val_split_idx:]

    print(f"\nVeri setleri bölündü:")
    print(f"  Eğitim seti boyutu: {len(train_df)} ({len(train_df)/num_total_samples*100:.1f}%)")
    print(f"  Doğrulama seti boyutu: {len(val_df)} ({len(val_df)/num_total_samples*100:.1f}%)")
    print(f"  Test seti boyutu: {len(test_df)} ({len(test_df)/num_total_samples*100:.1f}%)")

    # 3. Veri Ölçeklendirme (StandardScaler)
    # Ölçekleyiciyi sadece eğitim verisi üzerinde eğitelim
    scaler = StandardScaler()
    scaled_train_data = scaler.fit_transform(train_df) # Eğit ve dönüştür

    # Aynı ölçekleyiciyi doğrulama ve test verisine de uygulayalım
    scaled_val_data = scaler.transform(val_df) # Sadece dönüştür
    scaled_test_data = scaler.transform(test_df) # Sadece dönüştür

    print("\nVeri ölçeklendirme tamamlandı (StandardScaler kullanıldı).")
    print(f"  Ölçeklenmiş eğitim verisi şekli: {scaled_train_data.shape}")

    # Ölçeklenmiş verileri tekrar DataFrame'e dönüştürmek isteğe bağlıdır,
    # genellikle pencereleme için NumPy dizileri yeterlidir.
    # scaled_train_df = pd.DataFrame(scaled_train_data, columns=feature_columns, index=train_df.index)
    # scaled_val_df = pd.DataFrame(scaled_val_data, columns=feature_columns, index=val_df.index)
    # scaled_test_df = pd.DataFrame(scaled_test_data, columns=feature_columns, index=test_df.index)

    # 4. Zaman Dizisi Pencereleri Oluşturma
    # Autoformer için tipik sequence_length (seq_len) ve prediction_length (pred_len) değerleri
    # ETTm1 için seq_len=96 (1 gün), pred_len=[96, 192, 336, 720] (1, 2, 3.5, 7.5 gün)
    # Bu örnek için seq_len=96 ve pred_len=96 (1 günlük geçmişle 1 günlük tahmin) seçelim.

    global sequence_length, prediction_length # Diğer hücrelerde kullanmak için
    sequence_length = 96  # Modele girdi olarak verilecek geçmiş zaman adımı sayısı
    prediction_length = 96 # Modelin tahmin edeceği gelecek zaman adımı sayısı

    def create_sequences(input_data, seq_len, pred_len, target_col_index=-1):
        """
        Verilen zaman serisi verisinden girdi (X) ve hedef (y) dizileri oluşturur.
        input_data: Ölçeklendirilmiş zaman serisi verisi (NumPy array).
        seq_len: Girdi dizisinin uzunluğu.
        pred_len: Hedef dizisinin uzunluğu (tahmin ufku).
        target_col_index: Hedef değişkenin input_data içindeki sütun indeksi.
                          Eğer tüm sütunlar hedefse (multi-target), bu farklı yönetilebilir.
                          Autoformer genellikle çok değişkenli tahmin yapar, yani tüm sütunlar hedef olabilir.
                          Bu örnekte, tüm özellikleri tahmin etmeye çalışalım.
                          Ya da sadece 'OT'yi (son sütun) tahmin edelim.
                          Autoformer makalesinde, ETT'de tüm değişkenler tahmin edilir (multivariate forecast).
        """
        X = []
        y = []
        num_samples = len(input_data) - seq_len - pred_len + 1
        if num_samples <= 0:
            print(f"Uyarı: Yetersiz veri (uzunluk {len(input_data)}) ile {seq_len} seq_len ve {pred_len} pred_len için dizi oluşturulamıyor.")
            return np.array(X), np.array(y)

        for i in range(num_samples):
            X.append(input_data[i : i + seq_len]) # Geçmiş seq_len kadar veri
            # Hedef, girdi dizisinin sonundan sonraki pred_len kadar veri
            # Autoformer çok değişkenli tahmin yaptığı için y tüm özellikleri içerebilir.
            y.append(input_data[i + seq_len : i + seq_len + pred_len])
            # Eğer sadece tek bir hedef (örn: OT) tahmin edilecekse:
            # y.append(input_data[i + seq_len : i + seq_len + pred_len, target_col_index])

        return np.array(X), np.array(y)

    # Hedef değişkenin indeksi (OT son sütundu)
    # feature_columns listesinde 'OT'nin indeksini bulalım
    target_idx = feature_columns.index(target_column) if target_column in feature_columns else -1

    print(f"\nZaman dizisi pencereleri oluşturuluyor (seq_len={sequence_length}, pred_len={prediction_length})...")

    # X_train, y_train vb. global yapalım
    global X_train, y_train, X_val, y_val, X_test, y_test, scaler_g # scaler'ı da global yapalım

    X_train, y_train = create_sequences(scaled_train_data, sequence_length, prediction_length, target_idx)
    X_val, y_val = create_sequences(scaled_val_data, sequence_length, prediction_length, target_idx)
    X_test, y_test = create_sequences(scaled_test_data, sequence_length, prediction_length, target_idx)
    scaler_g = scaler # Ölçekleyiciyi global yapalım ki tahminleri geri ölçeklemek için kullanılabilsin

    print(f"Oluşturulan dizilerin şekilleri:")
    print(f"  X_train şekli: {X_train.shape}, y_train şekli: {y_train.shape}")
    print(f"  X_val şekli: {X_val.shape}, y_val şekli: {y_val.shape}")
    print(f"  X_test şekli: {X_test.shape}, y_test şekli: {y_test.shape}")

    # y_test'in ölçeklenmemiş halini de saklayalım (performans metriklerini orijinal ölçekte görmek için)
    # Bunun için orijinal test_df'ten ilgili kısmı almamız gerekir.
    # Bu biraz daha karmaşık, çünkü y_test'in başlangıç ve bitiş indekslerini orijinal df'te bulmalıyız.
    # Şimdilik, tahminleri yaptıktan sonra scaled_y_pred'i geri ölçekleyerek orijinal y_test ile karşılaştıracağız.

    if X_train.size == 0 or X_val.size == 0 or X_test.size == 0:
        print("UYARI: Oluşturulan eğitim, doğrulama veya test dizilerinden en az biri boş!")
        print("Bu durum, veri seti boyutunun seçilen sequence_length ve prediction_length için yetersiz olmasından kaynaklanabilir.")
        print("Lütfen veri bölme oranlarını veya sequence/prediction uzunluklarını gözden geçirin.")

    print("\nVeri ön işleme (Hücre 4) tamamlandı.")

else:
    print("Hata: 'df' DataFrame'i veya 'target_column' önceki hücrede doğru şekilde tanımlanmamış.")
    # Hata durumunda değişkenleri None olarak ayarlayalım ki sonraki hücreler kontrollü hata versin
    X_train, y_train, X_val, y_val, X_test, y_test, scaler_g = [None]*7
```

```python
# Hücre 5: Autoformer Modelinin Ortama Eklenmesi (Resmi Repo - ZIP İndirme)

import sys
import os
import shutil # Klasör işlemleri için

# Resmi Autoformer reposu ve branch adı (genellikle 'main' veya 'master')
# thuml/Autoformer için ana branch 'main' gibi görünüyor.
autoformer_official_repo_owner = "thuml"
autoformer_official_repo_name = "Autoformer"
branch_name = "main" # Veya 'master' olabilir, repo kontrol edilmeli

zip_download_url = f"https://github.com/{autoformer_official_repo_owner}/{autoformer_official_repo_name}/archive/refs/heads/{branch_name}.zip"
zip_file_name = f"{branch_name}.zip"
extracted_folder_name_from_zip = f"{autoformer_official_repo_name}-{branch_name}" # Örn: Autoformer-main
target_repo_folder_name = "autoformer_official_code" # Kodları bu klasöre taşıyacağız

# 1. Mevcut olabilecek eski klasörleri/dosyaları temizle
for item_to_remove in [target_repo_folder_name, extracted_folder_name_from_zip, zip_file_name]:
    if os.path.exists(item_to_remove):
        print(f"'{item_to_remove}' zaten var, içeriği siliniyor...")
        if os.path.isdir(item_to_remove):
            shutil.rmtree(item_to_remove, ignore_errors=True)
        else:
            try:
                os.remove(item_to_remove)
            except OSError:
                pass
        # print(f"'{item_to_remove}' silindi veya silinmeye çalışıldı.")

print(f"\n'{autoformer_official_repo_name}' reposunu ZIP olarak indirmek için işlemler başlatılıyor...")
files_obtained_from_zip = False

try:
    # 2. GitHub Reposunun ZIP Olarak İndirilmesi
    print(f"  '{zip_download_url}' adresinden '{zip_file_name}' olarak ZIP indiriliyor (wget çıktıları gösterilecek)...")
    # wget komutundan -q parametresini kaldırarak daha fazla çıktı alalım
    wget_status = os.system(f"wget -O {zip_file_name} \"{zip_download_url}\"") # URL tırnak içinde

    if wget_status == 0 and os.path.exists(zip_file_name):
        print(f"\n  '{zip_file_name}' başarıyla indirildi.")

        # 3. ZIP Dosyasını Açma
        print(f"  '{zip_file_name}' açılıyor...")
        unzip_status = os.system(f"unzip -o {zip_file_name}") # -o: overwrite, -q çıkarıldı

        if unzip_status == 0: # Başarılı unzip
            print(f"  '{zip_file_name}' başarıyla açıldı.")

            # 4. Klasörü Yeniden Adlandırma (eğer gerekiyorsa)
            if os.path.exists(extracted_folder_name_from_zip):
                print(f"  Açılan '{extracted_folder_name_from_zip}' klasörü '{target_repo_folder_name}' olarak yeniden adlandırılıyor...")
                if os.path.exists(target_repo_folder_name): # Hedef zaten varsa sil
                    shutil.rmtree(target_repo_folder_name, ignore_errors=True)
                shutil.move(extracted_folder_name_from_zip, target_repo_folder_name)

                if os.path.exists(target_repo_folder_name):
                    print(f"  Klasör başarıyla '{target_repo_folder_name}' olarak yeniden adlandırıldı.")
                    files_obtained_from_zip = True
                else:
                    print(f"  !!! HATA: Yeniden adlandırma sonrası '{target_repo_folder_name}' bulunamadı.")
            else: # Bazen unzip doğrudan repo adına bir klasör oluşturabilir, -main/master olmadan.
                if os.path.exists(autoformer_official_repo_name) and \
                   not os.path.exists(target_repo_folder_name) and \
                   autoformer_official_repo_name != target_repo_folder_name :
                    print(f"  Açılan klasör '{autoformer_official_repo_name}' olarak bulundu, '{target_repo_folder_name}' olarak yeniden adlandırılıyor...")
                    shutil.move(autoformer_official_repo_name, target_repo_folder_name)
                    if os.path.exists(target_repo_folder_name):
                        print(f"  Klasör başarıyla '{target_repo_folder_name}' olarak yeniden adlandırıldı.")
                        files_obtained_from_zip = True
                    else:
                        print(f"  !!! HATA: Yeniden adlandırma sonrası '{target_repo_folder_name}' bulunamadı.")
                elif os.path.exists(target_repo_folder_name): # Eğer zaten doğru isimde açıldıysa
                     print(f"  Açılan klasör zaten '{target_repo_folder_name}' adında.")
                     files_obtained_from_zip = True
                else:
                    print(f"  !!! HATA: Beklenen açılmış klasör ('{extracted_folder_name_from_zip}' veya '{autoformer_official_repo_name}') bulunamadı.")
                    print("  Mevcut dosyalar/klasörler (unzip sonrası):")
                    !ls -l
        else:
            print(f"  !!! HATA: ZIP açma işlemi başarısız oldu (unzip return code: {unzip_status}).")
            print("  Lütfen unzip komutunun üstteki çıktılarını kontrol edin.")
    else:
        print(f"  !!! HATA: ZIP indirme işlemi başarısız oldu (wget return code: {wget_status}).")
        print(f"  Lütfen wget komutunun üstteki çıktılarını (HTTP hataları vb.) kontrol edin.")

    if os.path.exists(zip_file_name): # Geçici zip dosyasını her durumda sil
        os.remove(zip_file_name)
        # print(f"  '{zip_file_name}' geçici dosyası silindi.")

except Exception as e_main_zip:
    print(f"!!! ZIP indirme/işleme ana bloğunda bir HATA oluştu: {e_main_zip}")
    import traceback
    print(traceback.format_exc())

# 5. Başarı Kontrolü ve Sistem Yoluna Ekleme
expected_model_file = os.path.join(target_repo_folder_name, "models", "Autoformer.py")
if files_obtained_from_zip and os.path.exists(expected_model_file):
    print(f"\nKaynak kodları başarıyla '{target_repo_folder_name}' klasörüne alındı ve '{expected_model_file}' bulundu.")

    repo_abs_path = os.path.abspath(target_repo_folder_name)
    if repo_abs_path not in sys.path:
        sys.path.insert(0, repo_abs_path) # sys.path'in başına ekleyerek öncelik verelim
        print(f"'{repo_abs_path}' yolu sys.path'e eklendi.")
    else:
        print(f"'{repo_abs_path}' yolu zaten sys.path'te mevcut.")

    print("\nAutoformer için ortam hazırlığı (kaynak kodları alma) tamamlandı.")
    print("Bir sonraki hücrede (Hücre 6) model sınıfını import etmeyi ve modeli yapılandırmayı deneyebilirsiniz.")
    print("UYARI: Bu modelin (thuml/Autoformer) PyTorch < 2.0 ve NumPy < 2.0 için yazıldığı unutulmamalıdır.")
    print("Mevcut PyTorch/NumPy versiyonlarımızla (örn: PyTorch 2.6.0, NumPy 2.0.2) uyumluluk sorunları yaşanması MUHTEMELDİR.")
    print("Bu sorunları Hücre 6'da (Model Tanımlama) ele almaya çalışacağız.")

    # Bu aşamada import testi yapmayalım, onu Hücre 6'ya bırakalım ki hataları orada görelim.
    # try:
    #     from models.Autoformer import Model as AutoformerModel
    #     print("\nTest: 'models.Autoformer.Model' başarıyla import edildi!")
    # except Exception as e_imp_test:
    #     print(f"\nUYARI: Model import testi Hücre 5'te başarısız oldu: {e_imp_test}")
    #     print("Bu, Hücre 6'da ele alınacak uyumluluk sorunlarına işaret ediyor olabilir.")

else:
    print(f"\n!!! HATA: Autoformer kaynak kodları '{target_repo_folder_name}' klasörüne alınamadı veya temel model dosyası ('{expected_model_file}') bulunamadı.")
    print("Lütfen adımları ve olası hata mesajlarını kontrol edin.")
    print(f"Son olarak '{target_repo_folder_name}' klasörünün durumu (eğer oluştuysa):")
    if os.path.exists(target_repo_folder_name):
        print("Klasör var, içeriği:")
        !ls -l {target_repo_folder_name}
        if os.path.exists(os.path.join(target_repo_folder_name, 'models')):
            print(f"\n'{target_repo_folder_name}/models/' klasörünün içeriği:")
            !ls -l {os.path.join(target_repo_folder_name, 'models')}
    else:
        print(f"'{target_repo_folder_name}' klasörü diskte bulunmuyor.")
```

            * `enc_in`: 7 (giriş özellik sayısı, ETTm1'deki sütun sayısı)
            * `dec_in`: 7 (dekoder girişi özellik sayısı, genellikle `enc_in` ile aynı)
            * `c_out`: 7 (çıkış özellik sayısı, çok değişkenli tahmin için `enc_in` ile aynı)
            * `d_model`: Modelin gizli katman boyutu (örn: 512)
            * `n_heads`: Çoklu dikkat (multi-head attention) başlık sayısı (örn: 8)
            * `e_layers`: Enkoder katman sayısı (örn: 2)
            * `d_layers`: Dekoder katman sayısı (örn: 1)
            * `d_ff`: Feed-forward katmanının iç boyutu (örn: 2048)
            * `dropout`: Dropout oranı (örn: 0.05)
            * `activation`: `'gelu'`
            * `output_attention`: Autoformer'da bu genellikle `False` olur, çünkü dikkat mekanizması farklıdır.
            * `embed`: Zaman özelliği gömme türü (örn: `'timeF'` - frekansa göre Fourier özellikleri)
            * `freq`: Veri frekansı (ETTm1 için `'t'` - 15 dakikalık, veya Autoformer kodunun beklediği formatta, örn: `'15t'`). Bu, zaman gömme katmanı için önemlidir.
            * `factor`: Autoformer'ın otomatik korelasyon mekanizmasındaki örnekleme faktörü (örn: 3 veya 5).
    3.  **Model Örneği Oluşturma:**
        * Tanımlanan yapılandırma (config) ile `AutoformerModelThuml(config)` şeklinde modelin bir örneği (`model_autoformer`) oluşturulmaya çalışılır.
    4.  **Modeli Cihaza Taşıma:**
        * Oluşturulan model, `model_autoformer.to(device)` ile GPU'ya (veya CPU'ya) taşınır.
    5.  **Hata Ayıklama (Gerekirse):**
        * Bu aşamada `ImportError`, `AttributeError`, `TypeError` gibi hatalar alınması beklenir. Örneğin, eski PyTorch versiyonlarındaki `torch.qr` gibi fonksiyonlar yenilerinde farklı modüllerde olabilir veya NumPy operasyonları değişmiş olabilir.
        * Alınan hatalara göre, klonlanan `autoformer_official_code` içindeki ilgili Python dosyalarında (örn: `models/Autoformer.py`, `layers/AutoCorrelation.py`, `layers/Embed.py` vb.) küçük, hedefe yönelik düzeltmeler yapılması gerekebilir. Bu, hatanın kaynağını bulmayı ve güncel API'lere göre kodu adapte etmeyi içerir. Bu adım interaktif ve zaman alıcı olabilir.

* **Uygulama Detayları/Sonuçlar (Beklenen):**
    * İdeal durumda, gerekli düzeltmeler yapıldıktan sonra Autoformer modelinin bir örneğinin başarıyla oluşturulması ve GPU'ya taşınması.
    * Modelin yapısının (`print(model_autoformer)`) ve parametre sayısının incelenebilir hale gelmesi.

```python
# Hücre 6: Autoformer Modelinin Tanımlanması ve Yapılandırılması

# sys ve os zaten Hücre 5'te import edilmiş olmalı ve sys.path ayarlanmış olmalı.
# device değişkeni Hücre 2'de tanımlanmıştı.
# X_train (ve dolayısıyla özellik sayısı), sequence_length, prediction_length Hücre 4'te global yapılmıştı.

# Önce model sınıfını import etmeyi deneyelim
# sys.path'e 'autoformer_official_code' klasörünün eklendiğini varsayıyoruz (Hücre 5'ten)
try:
    from models.Autoformer import Model as AutoformerModelThuml
    print("'models.Autoformer.Model' başarıyla AutoformerModelThuml olarak import edildi.")

    # Autoformer için yapılandırma (config) oluşturalım
    # Bu değerler ETTm1 veri seti ve Autoformer makalesindeki ayarlara göre örneklendirilmiştir.
    # Bunları kendi veri setinize ve deneylerinize göre ayarlamanız gerekebilir.

    class Configs:
        pass # Basit bir sınıf veya dotdict kullanabiliriz.
             # thuml/Autoformer utils/tools.py içinde dotdict var ama onu import etmeden de yapabiliriz.
             # Ya da doğrudan bir Python dictionary kullanıp model __init__'ine **kwargs ile geçebiliriz.
             # Modelin __init__ metodunun nasıl tanımlandığına bakalım:
             # def __init__(self, configs): şeklinde ise, config objesi bekliyor.

    configs = Configs() # Boş bir obje oluşturalım ve özelliklerini atayalım

    # Veri ve Tahmin Uzunlukları (Hücre 4'ten)
    configs.seq_len = sequence_length      # 96 (girdi dizisi uzunluğu)
    configs.label_len = sequence_length // 2 # 48 (dekoder için başlangıç token'ı, Autoformer'da yaygın bir ayar)
    configs.pred_len = prediction_length   # 96 (tahmin ufku)

    # Model Mimarisi Parametreleri
    configs.enc_in = X_train.shape[-1]     # Giriş özellik sayısı (ETTm1 için 7)
    configs.dec_in = X_train.shape[-1]     # Dekoder giriş özellik sayısı (genellikle enc_in ile aynı)
    configs.c_out = X_train.shape[-1]      # Çıkış özellik sayısı (çok değişkenli tahmin için enc_in ile aynı)

    configs.d_model = 512                  # Modelin gizli katman boyutu
    configs.n_heads = 8                    # Çoklu dikkat başlık sayısı
    configs.e_layers = 2                   # Enkoder katman sayısı
    configs.d_layers = 1                   # Dekoder katman sayısı
    configs.d_ff = 2048                    # Feed-forward katmanının iç boyutu
    configs.dropout = 0.05                 # Dropout oranı
    configs.activation = 'gelu'            # Aktivasyon fonksiyonu

    configs.output_attention = False       # Autoformer genellikle bunu False kullanır
    configs.embed = 'timeF'                # Zaman özelliği gömme türü ('timeF', 'fixed', 'learned')
                                           # ETT için 'timeF' yaygındır.
    configs.freq = 't'                     # Veri frekansı (ETTm1 15 dakikalık olduğu için 't' veya '15t')
                                           # Orijinal kod 'h' (saatlik) veya 't' (dakikalık) gibi bekleyebilir.
                                           # ETTm1 için 't' (15min) veya '15min' olabilir, kodun embed.py'sine bakmak lazım.
                                           # thuml/Autoformer data_parser.py'de 'freq' için 't' kullanılıyor ETTm için.

    # Autoformer'a özgü parametreler
    configs.factor = 3                     # Auto-correlation mekanizmasındaki örnekleme faktörü (1, 3, 5 gibi)
    configs.distil = True                  # Enkoderde distillation kullanılsın mı (True)
    configs.moving_avg = 25                # Ayrıştırma için hareketli ortalama pencere boyutu (tek sayı önerilir)

    # Diğer olası config parametreleri (modelin __init__ metoduna göre eklenebilir):
    # configs.num_workers = 0 (DataLoader için)
    # configs.patience = 3 (Early stopping için)
    # configs.learning_rate = 0.0001
    # configs.lradj = 'type1' (Öğrenme oranı ayarlama stratejisi)
    # configs.use_amp = False (Karışık hassasiyetli eğitim)
    # configs.inverse = True (Tahminleri geri ölçekleme) - Bu eğitim sonrası için.

    print("\nAutoformer için yapılandırma parametreleri:")
    for attr, value in configs.__dict__.items():
        print(f"  {attr}: {value}")

    # Model örneğini oluşturalım
    print("\nAutoformer modeli örneği oluşturuluyor...")
    # model_autoformer'ı global yapalım
    global model_autoformer
    model_autoformer = AutoformerModelThuml(configs) # Config objesini modele ver
    model_autoformer.to(device) # Modeli GPU'ya taşı

    print(f"\nAutoformer modeli başarıyla oluşturuldu ve '{device}' cihazına taşındı.")

    # Modelin yapısını veya parametre sayısını kontrol edebiliriz (isteğe bağlı)
    # print(model_autoformer)
    total_params_autoformer = sum(p.numel() for p in model_autoformer.parameters())
    trainable_params_autoformer = sum(p.numel() for p in model_autoformer.parameters() if p.requires_grad)
    print(f"Autoformer modelindeki toplam parametre sayısı: {total_params_autoformer:,}")
    print(f"Autoformer modelindeki eğitilebilir parametre sayısı: {trainable_params_autoformer:,}")
    print("\nUYARI: Eğer bu hücrede PyTorch veya NumPy API uyumsuzluklarından kaynaklanan hatalar alırsanız,")
    print("klonlanan 'autoformer_official_code' içindeki model dosyalarında küçük güncellemeler yapmanız gerekebilir.")
    print("Bu, genellikle eski fonksiyon isimlerinin yenileriyle değiştirilmesini veya bazı operasyonların")
    print("güncel kütüphane versiyonlarına göre ayarlanmasını içerir.")

except ImportError as e_imp:
    print(f"\nHATA: Autoformer modeli import edilirken bir sorun oluştu: {e_imp}")
    print("Lütfen Hücre 5'teki klonlama ve path ayarlarının doğru olduğundan emin olun.")
    print("Ayrıca, klonlanan kodun içindeki 'models' klasöründe 'Autoformer.py' dosyasının bulunduğunu teyit edin.")
    import traceback
    print(traceback.format_exc())
    if 'model_autoformer' in globals(): del model_autoformer

except Exception as e_model_init:
    print(f"\nAutoformer modeli oluşturulurken veya yapılandırılırken bir HATA oluştu: {e_model_init}")
    print("Bu hata, genellikle model kodunun mevcut PyTorch/NumPy versiyonlarıyla uyumsuzluğundan kaynaklanır.")
    print("Hata mesajını dikkatlice inceleyerek, 'autoformer_official_code' içindeki ilgili dosyalarda")
    print("düzeltmeler yapmanız gerekebilir (örn: eski fonksiyon adlarını güncellemek).")
    import traceback
    print(traceback.format_exc())
    if 'model_autoformer' in globals(): del model_autoformer
```

### Hücre 6a: Autoformer Modeli İçin Gelişmiş Veri Hazırlığı (Zaman Özellikleri ve Dekoder Girdisi - Düzeltilmiş)

* **Amaç:**

Bu, zaman özelliklerinin çıkarılmasını, doğru şekilde ölçeklendirilmesini ve PyTorch `Dataset` ve `DataLoader`'ları ile eğitim için hazırlanmasını içerir.

* **Yapılan İşlemler:**
    1.  **Zaman Özelliklerini Çıkarma (`time_features`):**

        * Bu dizilerden PyTorch `TensorDataset`'leri ve ardından `DataLoader`'lar oluşturulur.

* **Uygulama Detayları/Sonuçlar (Beklenen):**
    * Zaman özellikleri doğru şekle (`(örnek_sayısı, özellik_sayısı)`) getirilerek `StandardScaler` ile hatasız bir şekilde ölçeklendirilir.
    * Autoformer girdi dizileri doğru şekillerde oluşturulur.
    * Model eğitimi için `train_loader_af`, `val_loader_af`, `test_loader_af` DataLoader'ları hazır hale gelir.

```python
# Hücre 6a: Autoformer Modeli İçin Gelişmiş Veri Hazırlığı (Transpoze Düzeltmesiyle)

# Gerekli importlar
from utils.timefeatures import time_features # Autoformer reposundan
from torch.utils.data import TensorDataset, DataLoader
from sklearn.preprocessing import StandardScaler
# numpy, pandas, torch zaten Hücre 2'de import edilmiş olmalı.

# Hücre 4 ve 6'dan gelen değişkenlerin varlığını kontrol edelim
if 'train_df' not in globals() or 'scaled_train_data' not in globals() or \
   'val_df' not in globals() or 'scaled_val_data' not in globals() or \
   'test_df' not in globals() or 'scaled_test_data' not in globals() or \
   'configs' not in globals() or 'scaler_g' not in globals() or \
   'sequence_length' not in globals() or 'prediction_length' not in globals():
    print("HATA: Önceki hücrelerden (Hücre 3, 4 ve/veya 6) gerekli değişkenler yüklenemedi.")
    print("Lütfen önceki hücrelerin doğru çalıştığından emin olun.")
    # Hata durumunda sonraki kodun çalışmaması için bir exception fırlatabiliriz veya değişkenleri None yaparız
    # For now, let's assume user will re-run if this happens.
else:
    print("Autoformer için gelişmiş veri hazırlığı başlıyor...")

    # 1. Zaman Özelliklerini Çıkarma ve Ölçeklendirme
    try:
        train_time_features_raw = time_features(train_df.index, freq=configs.freq)
        val_time_features_raw = time_features(val_df.index, freq=configs.freq)
        test_time_features_raw = time_features(test_df.index, freq=configs.freq)

        print(f"  Orijinal ham zaman özellikleri şekilleri - Train: {train_time_features_raw.shape}, Val: {val_time_features_raw.shape}, Test: {test_time_features_raw.shape}")

        # DÜZELTME: Matrisleri transpoze et (örnek_sayısı, özellik_sayısı) formatına getir
        train_time_features_raw_t = train_time_features_raw.T
        val_time_features_raw_t = val_time_features_raw.T
        test_time_features_raw_t = test_time_features_raw.T

        print(f"  Transpoze edilmiş ham zaman özellikleri şekilleri - Train: {train_time_features_raw_t.shape}, Val: {val_time_features_raw_t.shape}, Test: {test_time_features_raw_t.shape}")

        scaler_time_features = StandardScaler()
        scaled_train_time_features = scaler_time_features.fit_transform(train_time_features_raw_t)
        scaled_val_time_features = scaler_time_features.transform(val_time_features_raw_t)
        scaled_test_time_features = scaler_time_features.transform(test_time_features_raw_t)

        print(f"  Ölçeklenmiş zaman özellikleri şekilleri - Train: {scaled_train_time_features.shape}")
        num_time_features = scaled_train_time_features.shape[-1]
        print(f"  Tespit edilen zaman özelliği sayısı: {num_time_features}")

        # scaler_time_features_g'yi global yapalım
        global scaler_time_features_g
        scaler_time_features_g = scaler_time_features

        # 2. Autoformer Girdi Dizilerini Oluşturma Fonksiyonu (öncekiyle aynı)
        def create_autoformer_sequences(main_data, time_data, seq_len, label_len, pred_len):
            X_enc, X_mark_enc, X_dec, X_mark_dec, Y_true = [], [], [], [], []
            num_samples_possible = len(main_data) - seq_len - pred_len + 1
            if num_samples_possible <= 0:
                return (np.array(X_enc), np.array(X_mark_enc), np.array(X_dec), np.array(X_mark_dec), np.array(Y_true))
            for i in range(num_samples_possible):
                X_enc.append(main_data[i : i + seq_len])
                X_mark_enc.append(time_data[i : i + seq_len])
                seq_y_start_idx = i + seq_len - label_len
                seq_y_end_idx = i + seq_len + pred_len
                seq_y_slice = main_data[seq_y_start_idx : seq_y_end_idx]
                seq_y_mark_slice = time_data[seq_y_start_idx : seq_y_end_idx]
                x_dec_history_part = seq_y_slice[:label_len]
                num_features = main_data.shape[-1]
                x_dec_future_padding = np.zeros((pred_len, num_features))
                x_dec_combined = np.concatenate([x_dec_history_part, x_dec_future_padding], axis=0)
                X_dec.append(x_dec_combined)
                X_mark_dec.append(seq_y_mark_slice)
                Y_true.append(seq_y_slice[label_len:])
            return (np.array(X_enc), np.array(X_mark_enc), np.array(X_dec), np.array(X_mark_dec), np.array(Y_true))

        print(f"\nAutoformer girdi dizileri oluşturuluyor (seq_len={configs.seq_len}, label_len={configs.label_len}, pred_len={configs.pred_len})...")

        global X_train_af, X_mark_train_af, X_dec_train_af, X_mark_dec_train_af, y_train_af
        global X_val_af, X_mark_val_af, X_dec_val_af, X_mark_dec_val_af, y_val_af
        global X_test_af, X_mark_test_af, X_dec_test_af, X_mark_dec_test_af, y_test_af

        X_train_af, X_mark_train_af, X_dec_train_af, X_mark_dec_train_af, y_train_af = \
            create_autoformer_sequences(scaled_train_data, scaled_train_time_features, configs.seq_len, configs.label_len, configs.pred_len)
        X_val_af, X_mark_val_af, X_dec_val_af, X_mark_dec_val_af, y_val_af = \
            create_autoformer_sequences(scaled_val_data, scaled_val_time_features, configs.seq_len, configs.label_len, configs.pred_len)
        X_test_af, X_mark_test_af, X_dec_test_af, X_mark_dec_test_af, y_test_af = \
            create_autoformer_sequences(scaled_test_data, scaled_test_time_features, configs.seq_len, configs.label_len, configs.pred_len)

        print("\nOluşturulan Autoformer girdi/çıktı dizilerinin şekilleri:")
        print(f"  X_train_af: {X_train_af.shape}, X_mark_train_af: {X_mark_train_af.shape}, X_dec_train_af: {X_dec_train_af.shape}, X_mark_dec_train_af: {X_mark_dec_train_af.shape}, y_train_af: {y_train_af.shape}")
        # Diğer val ve test şekillerini de yazdırabiliriz ama genellikle train ile benzer olur (örnek sayısı hariç)

        # 3. PyTorch Dataset ve DataLoader Hazırlığı
        batch_size_autoformer = 32

        train_dataset_af = TensorDataset(
            torch.from_numpy(X_train_af).float(), torch.from_numpy(X_mark_train_af).float(),
            torch.from_numpy(X_dec_train_af).float(), torch.from_numpy(X_mark_dec_train_af).float(),
            torch.from_numpy(y_train_af).float()
        )
        # Diğer datasetler için de benzeri...
        val_dataset_af = TensorDataset(
            torch.from_numpy(X_val_af).float(), torch.from_numpy(X_mark_val_af).float(),
            torch.from_numpy(X_dec_val_af).float(), torch.from_numpy(X_mark_dec_val_af).float(),
            torch.from_numpy(y_val_af).float()
        )
        test_dataset_af = TensorDataset(
            torch.from_numpy(X_test_af).float(), torch.from_numpy(X_mark_test_af).float(),
            torch.from_numpy(X_dec_test_af).float(), torch.from_numpy(X_mark_dec_test_af).float(),
            torch.from_numpy(y_test_af).float()
        )

        global train_loader_af, val_loader_af, test_loader_af
        train_loader_af = DataLoader(train_dataset_af, batch_size=batch_size_autoformer, shuffle=True)
        val_loader_af = DataLoader(val_dataset_af, batch_size=batch_size_autoformer, shuffle=False)
        test_loader_af = DataLoader(test_dataset_af, batch_size=batch_size_autoformer, shuffle=False)

        print(f"\nPyTorch DataLoader'lar oluşturuldu (batch_size={batch_size_autoformer}).")
        print(f"  Train loader örnek sayısı: {len(train_loader_af.dataset)} -> {len(train_loader_af)} batch")

        print("\nAutoformer için gelişmiş veri hazırlığı (Hücre 6a) tamamlandı.")

    except Exception as e_prep:
        print(f"\nVeri hazırlığı sırasında bir HATA oluştu: {e_prep}")
        import traceback
        print(traceback.format_exc())
        # Hata durumunda değişkenleri temizleyelim
        vars_to_clear = ['X_train_af', 'X_mark_train_af', 'X_dec_train_af', 'X_mark_dec_train_af', 'y_train_af',
                         'X_val_af', 'X_mark_val_af', 'X_dec_val_af', 'X_mark_dec_val_af', 'y_val_af',
                         'X_test_af', 'X_mark_test_af', 'X_dec_test_af', 'X_mark_dec_test_af', 'y_test_af',
                         'scaler_time_features_g', 'train_loader_af', 'val_loader_af', 'test_loader_af']
        for var_name in vars_to_clear:
            if var_name in globals(): del globals()[var_name]
```

### Hücre 7 (Autoformer): Modelin Eğitilmesi (Gelişmiş Metriklerle Güncellenmiş PyTorch Eğitim Döngüsü)

* **Amaç:**
    Hazırlanan veri yükleyicilerini ve tanımlanan Autoformer modelini kullanarak modeli eğitmek. Eğitim sırasında kayıp (loss) değerlerini ve proje isterlerinde belirtilen **tüm regresyon metriklerini (MSE, MAE, RMSE, MAPE, R-Squared)** doğrulama seti üzerinde her epoch sonunda izlemek, en iyi modeli kaydetmek.

* **Yapılan İşlemler:**
    1.  **Eğitim Hiperparametreleri:** Epoch sayısı, öğrenme oranı, erken durdurma sabrı belirlenir.
    2.  **Kayıp Fonksiyonu:** `torch.nn.MSELoss()` kullanılır.
    3.  **Optimizatör:** `torch.optim.Adam` kullanılır.
    4.  **`compute_regression_metrics` Fonksiyonu:**
        * Bir epoch'taki tüm doğrulama seti tahminlerini ve gerçek değerlerini (NumPy dizileri olarak) alır.
        * `sklearn.metrics` kullanarak MAE, RMSE (MSE'den türetilir), MAPE ve R-Squared (`r2_score`) hesaplar. MSE zaten epoch kaybı olarak mevcuttur.
        * Bu metrikleri içeren bir sözlük döndürür. MAPE hesaplanırken sıfıra bölme hatalarına karşı dikkatli olunur.
    5.  **Eğitim ve Doğrulama Döngüsü:**
        * Her epoch için:
            * **Eğitim Adımı:** Model eğitim moduna alınır, `train_loader_af` üzerinden batch'ler işlenir, kayıp hesaplanır, gradyanlar güncellenir. Ortalama eğitim kaybı loglanır.
            * **Doğrulama Adımı:** Model değerlendirme moduna alınır. `val_loader_af` üzerinden tüm doğrulama verisi için tahminler ve gerçek değerler toplanır. Epoch sonunda toplanan bu değerler kullanılarak ortalama doğrulama kaybı (MSE) ve `compute_regression_metrics` ile diğer tüm regresyon metrikleri hesaplanır ve loglanır.
        * **En İyi Modeli Kaydetme ve Erken Durdurma:** Doğrulama kaybına (veya istenirse başka bir metriğe) göre en iyi model kaydedilir ve erken durdurma uygulanır.
    6.  Toplam eğitim süresi ölçülür ve en iyi modelin ağırlıkları eğitim sonunda modele yüklenir.
    7.  Eğitim ve doğrulama kayıpları ile hesaplanan doğrulama metrikleri, daha sonra grafik çizdirmek üzere listelerde saklanır.

* **Uygulama Detayları/Sonuçlar (Beklenen):**
    * Modelin eğitilmesi.
    * Her epoch sonunda eğitim kaybının yanı sıra, doğrulama kaybı ve **tüm istenen regresyon metriklerinin (MAE, RMSE, MAPE, R2)** yazdırılması.
    * En iyi modelin (en düşük doğrulama kaybına göre) kaydedilmesi.
    * `train_losses_af`, `val_losses_af` ve yeni oluşturulacak `val_metrics_history_af` listelerinin, Hücre 9'daki grafik çizimleri için veri içermesi.

```python
# Hücre 7 (Autoformer): Modelin Eğitilmesi (Gelişmiş Metriklerle Güncellenmiş PyTorch Eğitim Döngüsü)

import torch.optim as optim
import torch.nn as nn
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, mean_absolute_percentage_error
# Diğer gerekli importlar (torch, time, np, tqdm) önceki hücrelerden gelmiş olmalı.
from tqdm.auto import tqdm

# Gerekli değişkenlerin (model_autoformer, train_loader_af, val_loader_af, device, configs, scaler_g)
# önceki hücrelerde doğru şekilde tanımlandığını varsayıyoruz.
if 'model_autoformer' not in globals() or model_autoformer is None or \
   'train_loader_af' not in globals() or train_loader_af is None or \
   'val_loader_af' not in globals() or val_loader_af is None or \
   'device' not in globals() or 'configs' not in globals() or \
   'scaler_g' not in globals() or scaler_g is None: # scaler_g de önemli
    print("HATA: Model, veri yükleyiciler, ölçekleyici veya cihaz bilgisi önceki hücrelerden yüklenemedi.")
else:
    print("Autoformer modeli için eğitim döngüsü (gelişmiş metriklerle) başlatılıyor...")

    # Eğitim Hiperparametreleri
    num_epochs = 10
    learning_rate = 0.0001
    early_stopping_patience = 3
    best_val_loss = float('inf')
    epochs_no_improve = 0
    best_model_state_path = './autoformer_best_model_v2.pth' # Farklı bir isim verelim

    criterion = nn.MSELoss()
    optimizer = optim.Adam(model_autoformer.parameters(), lr=learning_rate)

    # Metrikleri ve kayıpları saklamak için global listeler
    global train_losses_af, val_losses_af, val_metrics_history_af
    train_losses_af = []
    val_losses_af = []
    val_metrics_history_af = [] # Her epoch'un doğrulama metriklerini tutacak (sözlük listesi)

    # Regresyon metriklerini hesaplama fonksiyonu
    def compute_regression_metrics(y_true_np, y_pred_np):
        # y_true_np ve y_pred_np (num_samples, pred_len, num_features) şeklinde olmalı
        # Metrikleri genellikle ana hedef değişken (örn: 'OT', son sütun) üzerinden hesaplarız
        # veya her özellik için ayrı ayrı hesaplayıp ortalamasını alabiliriz.
        # Şimdilik tüm özellikler üzerinden ortalama metrikleri veya ilk özellik için hesaplayalım.
        # Autoformer çok değişkenli tahmin yaptığı için, tüm çıktılar üzerinden hesaplayalım.
        # Boyutları (num_samples * pred_len, num_features) yapalım
        y_true_flat = y_true_np.reshape(-1, y_true_np.shape[-1])
        y_pred_flat = y_pred_np.reshape(-1, y_pred_np.shape[-1])

        mse = mean_squared_error(y_true_flat, y_pred_flat)
        mae = mean_absolute_error(y_true_flat, y_pred_flat)
        rmse = np.sqrt(mse)
        r2 = r2_score(y_true_flat, y_pred_flat)

        # MAPE için sıfıra bölme hatasını önlemek için dikkatli olalım
        # sklearn.metrics.mean_absolute_percentage_error bunu kendi içinde yönetir.
        # Ancak, MAPE genellikle %0-100 aralığında yorumlanır ve negatif değerler için anlamsız olabilir.
        # ETT verisindeki 'OT' genellikle pozitif.
        mape = mean_absolute_percentage_error(y_true_flat, y_pred_flat) * 100 # Yüzde olarak

        return {'mse': mse, 'mae': mae, 'rmse': rmse, 'r2': r2, 'mape': mape}

    print(f"Eğitim {num_epochs} epoch için başlatılıyor. Öğrenme Oranı: {learning_rate}")
    start_time_autoformer_train = time.time()

    for epoch in range(1, num_epochs + 1):
        model_autoformer.train()
        current_epoch_train_loss = 0.0
        train_iterator = tqdm(train_loader_af, desc=f"Epoch {epoch}/{num_epochs} [Eğitim]")

        for batch_data in train_iterator:
            batch_x_enc, batch_x_mark_enc, batch_x_dec, batch_x_mark_dec, batch_y_true = [d.to(device) for d in batch_data]
            optimizer.zero_grad()
            model_outputs = model_autoformer(batch_x_enc, batch_x_mark_enc, batch_x_dec, batch_x_mark_dec)
            if isinstance(model_outputs, tuple): model_outputs = model_outputs[0] # Eğer model birden fazla şey döndürüyorsa
            loss = criterion(model_outputs, batch_y_true)
            loss.backward()
            optimizer.step()
            current_epoch_train_loss += loss.item()
            train_iterator.set_postfix({'loss': loss.item()})
        avg_epoch_train_loss = current_epoch_train_loss / len(train_loader_af)
        train_losses_af.append(avg_epoch_train_loss)

        model_autoformer.eval()
        current_epoch_val_loss = 0.0
        all_val_preds_epoch = []
        all_val_true_epoch = []
        val_iterator = tqdm(val_loader_af, desc=f"Epoch {epoch}/{num_epochs} [Doğrulama]", leave=False)

        with torch.no_grad():
            for batch_data_val in val_iterator:
                val_x_enc, val_x_mark_enc, val_x_dec, val_x_mark_dec, val_y_true = [d.to(device) for d in batch_data_val]
                val_outputs = model_autoformer(val_x_enc, val_x_mark_enc, val_x_dec, val_x_mark_dec)
                if isinstance(val_outputs, tuple): val_outputs = val_outputs[0]
                val_loss = criterion(val_outputs, val_y_true)
                current_epoch_val_loss += val_loss.item()
                all_val_preds_epoch.append(val_outputs.cpu().numpy())
                all_val_true_epoch.append(val_y_true.cpu().numpy())
                val_iterator.set_postfix({'val_loss': val_loss.item()})

        avg_epoch_val_loss = current_epoch_val_loss / len(val_loader_af)
        val_losses_af.append(avg_epoch_val_loss)

        # Tüm doğrulama tahminlerini ve gerçeklerini birleştir
        y_pred_epoch_np = np.concatenate(all_val_preds_epoch, axis=0)
        y_true_epoch_np = np.concatenate(all_val_true_epoch, axis=0)

        # Metrikleri hesapla (ölçeklenmiş veriler üzerinden)
        # Tahminleri ve gerçek değerleri geri ölçeklendirmeden metrik hesaplamak,
        # modelin ölçeklenmiş uzaydaki performansını gösterir.
        # Raporlama için genellikle geri ölçeklendirilmiş değerler üzerinden metrikler verilir.
        # Şimdilik ölçeklenmiş üzerinden hesaplayalım, test setinde geri ölçeklendirmeyi yaparız.
        epoch_val_metrics = compute_regression_metrics(y_true_epoch_np, y_pred_epoch_np)
        epoch_val_metrics['val_loss'] = avg_epoch_val_loss # MSE'yi val_loss olarak da ekleyelim
        val_metrics_history_af.append(epoch_val_metrics)

        print(f"Epoch {epoch}/{num_epochs} -> Eğitim Kaybı: {avg_epoch_train_loss:.6f}, Doğrulama Kaybı (MSE): {avg_epoch_val_loss:.6f}")
        print(f"  Val MAE: {epoch_val_metrics['mae']:.6f}, Val RMSE: {epoch_val_metrics['rmse']:.6f}, Val MAPE: {epoch_val_metrics['mape']:.2f}%, Val R2: {epoch_val_metrics['r2']:.6f}")

        if avg_epoch_val_loss < best_val_loss:
            best_val_loss = avg_epoch_val_loss
            torch.save(model_autoformer.state_dict(), best_model_state_path)
            print(f"  Epoch {epoch}: Doğrulama kaybı iyileşti ({best_val_loss:.6f}), model kaydedildi -> {best_model_state_path}")
            epochs_no_improve = 0
        else:
            epochs_no_improve += 1
            print(f"  Epoch {epoch}: Doğrulama kaybı iyileşmedi. Sabır: {epochs_no_improve}/{early_stopping_patience}")
            if epochs_no_improve >= early_stopping_patience:
                print(f"Erken durdurma! {early_stopping_patience} epoch boyunca doğrulama kaybı iyileşmedi.")
                break
        print("-" * 70)

    # training_time_autoformer'ı global yapalım
    global training_time_autoformer
    end_time_autoformer_train = time.time()
    training_time_autoformer = end_time_autoformer_train - start_time_autoformer_train
    print(f"\nAutoformer eğitimi tamamlandı! Toplam eğitim süresi: {training_time_autoformer:.2f} saniye ({training_time_autoformer/60:.2f} dakika).")

    if os.path.exists(best_model_state_path):
        print(f"En iyi model '{best_model_state_path}' adresinden yükleniyor...")
        model_autoformer.load_state_dict(torch.load(best_model_state_path, map_location=device))
        print("En iyi model başarıyla yüklendi.")
    else:
        print(f"Uyarı: Kaydedilmiş en iyi model ('{best_model_state_path}') bulunamadı.")
```

### Hücre 8 (Autoformer): Eğitilmiş Modelin Test Seti Üzerinde Değerlendirilmesi ve Metrik Hesaplanması

* **Amaç:**
    Hücre 7'de eğitimi tamamlanan ve en iyi checkpoint'i yüklenmiş olan Autoformer modelinin, daha önce hiç görmediği test veri seti üzerindeki nihai tahmin performansını ölçmek. Proje isterlerinde belirtilen regresyon metriklerini (MSE, MAE, RMSE, MAPE, R-Squared) hesaplamak. Tahminlerin ve gerçek değerlerin orijinal ölçeklerine geri dönüştürülmesi önemlidir.

* **Yapılan İşlemler:**
    1.  Eğitilmiş `model_autoformer` değerlendirme moduna (`model.eval()`) alınır.
    2.  `test_loader_af` kullanılarak test seti üzerinde tahminler (`y_pred_test_list`) ve karşılık gelen gerçek değerler (`y_true_test_list`) toplanır. Modelin çıktısı, ölçeklendirilmiş değerlerdir.
    3.  Toplanan tüm tahminler ve gerçek değerler birleştirilerek NumPy dizilerine dönüştürülür.
    4.  Hem tahmin edilen değerler hem de gerçek değerler, Hücre 4'te eğitim verisi üzerinde eğitilen `scaler_g` (ana özellikler için `StandardScaler`) kullanılarak orijinal ölçeklerine geri dönüştürülür (`inverse_transform`). Bu işlem, tüm özellikler için yapılır.
    5.  **Önemli:** Metrikler genellikle ana hedef değişken üzerinden değerlendirilir. Bu çalışmada hedef değişkenimiz 'OT' (Oil Temperature) olduğu için, geri ölçeklendirilmiş tahminlerden ve gerçek değerlerden sadece 'OT' sütununa karşılık gelen değerler seçilir.
    6.  Seçilen 'OT' tahminleri ve gerçek 'OT' değerleri kullanılarak `sklearn.metrics` ile MSE, MAE, RMSE, MAPE ve R-Squared metrikleri hesaplanır.
    7.  Hesaplanan bu metrikler ekrana yazdırılır ve daha sonra raporlama için saklanır.

* **Uygulama Detayları/Sonuçlar (Beklenen):**
    * Test seti üzerinde tahminler yapılırken bir ilerleme çubuğu görülebilir.
    * Geri ölçeklendirilmiş (orijinal birimlerdeki) 'OT' tahminleri ve gerçek 'OT' değerleri üzerinden hesaplanan MSE, MAE, RMSE, MAPE ve R-Squared metrikleri ekrana yazdırılır.
    * Bu metrikler, modelin test verisi üzerindeki tahmin doğruluğu ve genelleme yeteneği hakkında nihai bir değerlendirme sunar. MAPE değeri, 'OT' değerlerinin aralığına bağlı olarak yorumlanmalıdır.

```python
# Hücre 8 (Autoformer): Eğitilmiş Modelin Test Seti Üzerinde Değerlendirilmesi ve Metrik Hesaplanması

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, mean_absolute_percentage_error
# Diğer gerekli importlar (torch, np, time, tqdm) önceki hücrelerden gelmiş olmalı

# Gerekli değişkenlerin (model_autoformer, test_loader_af, device, scaler_g, target_column, feature_columns, prediction_length)
# önceki hücrelerde doğru şekilde tanımlandığını varsayıyoruz.
if 'model_autoformer' not in globals() or model_autoformer is None or \
   'test_loader_af' not in globals() or test_loader_af is None or \
   'scaler_g' not in globals() or scaler_g is None or \
   'target_column' not in globals() or \
   'feature_columns' not in globals() or \
   'prediction_length' not in globals() or \
   'device' not in globals():
    print("HATA: Model, test veri yükleyici, ölçekleyici veya diğer gerekli yapılandırmalar yüklenemedi.")
else:
    print("Eğitilmiş Autoformer modeli test seti üzerinde değerlendiriliyor...")
    model_autoformer.eval() # Modeli değerlendirme moduna al

    y_true_test_list = []
    y_pred_test_list = []

    test_iterator_eval = tqdm(test_loader_af, desc="Test Seti Değerlendirme")
    with torch.no_grad():
        for batch_data_test in test_iterator_eval:
            test_x_enc, test_x_mark_enc, test_x_dec, test_x_mark_dec, test_y_true = [d.to(device) for d in batch_data_test]

            # Modelden tahminleri al
            test_outputs = model_autoformer(test_x_enc, test_x_mark_enc, test_x_dec, test_x_mark_dec)
            if isinstance(test_outputs, tuple): # Eğer model birden fazla şey döndürüyorsa (örn: attention)
                test_outputs = test_outputs[0]

            y_true_test_list.append(test_y_true.cpu().numpy())
            y_pred_test_list.append(test_outputs.cpu().numpy())

    # Tüm tahminleri ve gerçekleri birleştir
    y_true_test_np_scaled = np.concatenate(y_true_test_list, axis=0) # Şekil: (num_test_samples, pred_len, num_features)
    y_pred_test_np_scaled = np.concatenate(y_pred_test_list, axis=0)

    print(f"\nÖlçeklenmiş Test Tahminleri Şekli: {y_pred_test_np_scaled.shape}")
    print(f"Ölçeklenmiş Gerçek Test Değerleri Şekli: {y_true_test_np_scaled.shape}")

    # Tahminleri ve gerçek değerleri orijinal ölçeğe geri dönüştür
    # scaler_g.transform() 2D dizi bekler: (samples, features)
    # pred_len'i sample boyutuyla birleştirelim
    num_test_samples_eff = y_true_test_np_scaled.shape[0]
    num_features = y_true_test_np_scaled.shape[2]

    y_true_test_np_reshaped = y_true_test_np_scaled.reshape(num_test_samples_eff * prediction_length, num_features)
    y_pred_test_np_reshaped = y_pred_test_np_scaled.reshape(num_test_samples_eff * prediction_length, num_features)

    y_true_test_np_inversed = scaler_g.inverse_transform(y_true_test_np_reshaped)
    y_pred_test_np_inversed = scaler_g.inverse_transform(y_pred_test_np_reshaped)

    # Tekrar (num_test_samples, pred_len, num_features) şekline getirelim (isteğe bağlı)
    y_true_test_final = y_true_test_np_inversed.reshape(num_test_samples_eff, prediction_length, num_features)
    y_pred_test_final = y_pred_test_np_inversed.reshape(num_test_samples_eff, prediction_length, num_features)

    # Metrikleri hedef değişken 'OT' üzerinden hesaplayalım
    try:
        target_col_idx_eval = feature_columns.index(target_column)
    except ValueError:
        print(f"HATA: Hedef sütun '{target_column}' özellik listesinde bulunamadı. Metrikler hesaplanamıyor.")
        target_col_idx_eval = -1 # Veya hata ver

    if target_col_idx_eval != -1:
        ot_true = y_true_test_final[:, :, target_col_idx_eval].flatten() # Tüm tahmin ufkundaki tüm örnekler için
        ot_pred = y_pred_test_final[:, :, target_col_idx_eval].flatten()

        mse_test = mean_squared_error(ot_true, ot_pred)
        mae_test = mean_absolute_error(ot_true, ot_pred)
        rmse_test = np.sqrt(mse_test)
        r2_test = r2_score(ot_true, ot_pred)

        # MAPE için dikkatli olalım (sıfır değerler)
        # mean_absolute_percentage_error, y_true=0 olan yerleri hesaba katmaz veya uyarı verir.
        # Eğer ot_true negatif değerler içeriyorsa MAPE anlamsız olabilir. ETT 'OT' genellikle pozitiftir.
        try:
            mape_test = mean_absolute_percentage_error(ot_true, ot_pred) * 100
        except Exception as e_mape:
            print(f"MAPE hesaplanırken hata/uyarı: {e_mape}. MAPE 0 olarak ayarlandı.")
            mape_test = 0.0 # Veya np.nan

        print("\nAutoformer Test Seti Performans Metrikleri (Geri Ölçeklendirilmiş 'OT' Üzerinden):")
        print(f"  Test MSE: {mse_test:.6f}")
        print(f"  Test MAE: {mae_test:.6f}")
        print(f"  Test RMSE: {rmse_test:.6f}")
        print(f"  Test MAPE: {mape_test:.2f}%")
        print(f"  Test R-Squared (R2): {r2_test:.6f}")

        # Bu metrikleri de saklayalım
        global autoformer_final_test_metrics
        autoformer_final_test_metrics = {
            'mse': mse_test, 'mae': mae_test, 'rmse': rmse_test,
            'mape': mape_test, 'r2': r2_test
        }
    else:
        autoformer_final_test_metrics = None
        print("Hedef sütun indeksi bulunamadığı için metrikler hesaplanamadı.")
```

### Hücre 10 (Autoformer): Eğitim ve Doğrulama Kayıp/Metrik Grafiğinin Çizdirilmesi

* **Amaç:**
    Autoformer modelinin eğitim süreci boyunca performansının nasıl değiştiğini (kayıp ve anahtar regresyon metrikleri açısından) hem eğitim hem de doğrulama veri setleri üzerinde epoch bazında görselleştirmek.

* **Yapılan İşlemler:**
    1.  Hücre 7'deki eğitim döngüsü sırasında her epoch için kaydedilen eğitim kayıpları (`train_losses_af`), doğrulama kayıpları (`val_losses_af`) ve diğer doğrulama metrikleri (`val_metrics_history_af` içindeki MAE, RMSE, MAPE, R2) kullanılır.
    2.  `matplotlib.pyplot` kullanılarak:
        * **Birinci Grafik:** Epoch bazında "Eğitim Kaybı (MSE)" ve "Doğrulama Kaybı (MSE)" aynı eksen üzerinde çizdirilir.
        * **İkinci Grafik (veya daha fazla):** Epoch bazında önemli doğrulama metrikleri (örn: MAE, RMSE, R-Squared) çizdirilir. MAPE değeri çok büyükse grafiği domine edebilir, bu yüzden dikkatli olunmalı veya ayrı çizdirilmelidir.

* **Uygulama Detayları/Sonuçlar (Beklenen):**
    * Grafikler, modelin öğrenme eğrisini, eğitim ve doğrulama performansları arasındaki farkı ve aşırı öğrenmenin ne zaman başladığını gösterir.
    * Doğrulama metriklerinin epoch'lar boyunca seyri, modelin genelleme yeteneğinin nasıl değiştiği hakkında fikir verir.

```python
# Hücre 10 (Autoformer): Eğitim ve Doğrulama Kayıp/Metrik Grafiğinin Çizdirilmesi

import matplotlib.pyplot as plt
import pandas as pd # DataFrame oluşturmak için
import numpy as np

# Hücre 7'den train_losses_af, val_losses_af, val_metrics_history_af listelerinin geldiğini varsayıyoruz
if 'train_losses_af' in globals() and 'val_losses_af' in globals() and 'val_metrics_history_af' in globals() and \
   len(train_losses_af) > 0 and len(val_losses_af) > 0 and len(val_metrics_history_af) > 0 and \
   len(train_losses_af) == len(val_losses_af) and len(val_losses_af) == len(val_metrics_history_af) :

    epochs_range = range(1, len(train_losses_af) + 1)

    # --- Kayıp Grafiği ---
    plt.figure(figsize=(12, 6))
    plt.plot(epochs_range, train_losses_af, 'b-o', label='Eğitim Kaybı (MSE)')
    plt.plot(epochs_range, val_losses_af, 'r-s', label='Doğrulama Kaybı (MSE)')
    plt.title('Autoformer: Epoch Bazında Eğitim ve Doğrulama Kaybı (MSE)')
    plt.xlabel('Epoch')
    plt.ylabel('Ortalama Kare Hata (MSE)')
    plt.xticks(list(epochs_range)) # Her epoch için bir işaret
    plt.legend()
    plt.grid(True)
    plt.show()

    # --- Diğer Doğrulama Metrikleri Grafiği ---
    # val_metrics_history_af bir sözlük listesiydi. Bunu DataFrame'e çevirelim.
    val_metrics_df = pd.DataFrame(val_metrics_history_af)
    val_metrics_df['epoch'] = epochs_range # Epoch bilgisini ekleyelim

    # MAE ve RMSE
    plt.figure(figsize=(12, 6))
    plt.plot(val_metrics_df['epoch'], val_metrics_df['mae'], 'g-^', label='Doğrulama MAE')
    plt.plot(val_metrics_df['epoch'], val_metrics_df['rmse'], 'm-v', label='Doğrulama RMSE')
    plt.title('Autoformer: Epoch Bazında Doğrulama MAE ve RMSE')
    plt.xlabel('Epoch')
    plt.ylabel('Hata Değeri')
    plt.xticks(list(epochs_range))
    plt.legend()
    plt.grid(True)
    plt.show()

    # R-Squared ve MAPE (MAPE yüzdelik olduğu için ayrı veya ikinci y-ekseninde çizilebilir)
    fig, ax1_r2 = plt.subplots(figsize=(12, 6))

    color = 'tab:orange'
    ax1_r2.set_xlabel('Epoch')
    ax1_r2.set_ylabel('R-Squared', color=color)
    ax1_r2.plot(val_metrics_df['epoch'], val_metrics_df['r2'], color=color, marker='p', linestyle='-', label='Doğrulama R-Squared')
    ax1_r2.tick_params(axis='y', labelcolor=color)
    ax1_r2.grid(True, axis='y', linestyle=':', alpha=0.7) # Sadece R2 için grid

    ax2_mape = ax1_r2.twinx()  # Aynı x-eksenini paylaşan ikinci y-ekseni
    color = 'tab:brown'
    ax2_mape.set_ylabel('MAPE (%)', color=color)
    ax2_mape.plot(val_metrics_df['epoch'], val_metrics_df['mape'], color=color, marker='h', linestyle='--', label='Doğrulama MAPE (%)')
    ax2_mape.tick_params(axis='y', labelcolor=color)

    plt.title('Autoformer: Epoch Bazında Doğrulama R-Squared ve MAPE')
    fig.tight_layout()
    # Lejantları birleştir
    lines, labels = ax1_r2.get_legend_handles_labels()
    lines2, labels2 = ax2_mape.get_legend_handles_labels()
    ax2_mape.legend(lines + lines2, labels + labels2, loc='best')
    plt.xticks(list(epochs_range))
    plt.grid(True, axis='x') # X ekseni için ana grid
    plt.show()

else:
    print("Hata: Grafik çizdirmek için yeterli kayıp/metrik geçmişi bulunamadı.")
    print("Lütfen Hücre 7'deki eğitim döngüsünün doğru logları ürettiğinden emin olun.")
    if 'train_losses_af' in globals(): print(f"  Eğitim Kayıpları: {train_losses_af}")
    if 'val_losses_af' in globals(): print(f"  Doğrulama Kayıpları: {val_losses_af}")
    if 'val_metrics_history_af' in globals(): print(f"  Doğrulama Metrikleri Geçmişi: {val_metrics_history_af}")
```

### Hücre 11 (Autoformer): Eğitim ve Çıkarım Sürelerinin Hesaplanması

* **Amaç:**
    Autoformer modelinin hem eğitim süresini hem de eğitilmiş modelle yeni zaman serisi verileri üzerinde tahmin yapma (çıkarım) hızını belirlemek.

* **Yapılan İşlemler:**
    1.  **Eğitim Süresi:** Hücre 7'de model eğitimi sırasında ölçülen ve `training_time_autoformer` değişkeninde saklanan toplam eğitim süresi saniye ve dakika cinsinden yazdırılır.
    2.  **Örnek Başına Ortalama Çıkarım Süresi:**
        * Model (`model_autoformer` - en iyi checkpoint yüklendiği varsayılır) değerlendirme moduna alınır.
        * `test_loader_af` (Hücre 6a'da oluşturulan) kullanılır.
        * `torch.no_grad()` bloğu içinde, test veri yükleyicisindeki her bir yığın için modelden tahminler alınır ve yığının işlenme süresi ölçülür. **Burada dikkat edilmesi gereken, bir "örneğin" ne olduğudur. Her bir `test_loader_af` öğesi, `prediction_length` kadar zaman adımını içeren bir tahmini temsil eder. Bu yüzden, "örnek başına" süre, "tahmin penceresi başına" süre olarak yorumlanabilir.**
        * Toplam çıkarım süresi ve işlenen toplam tahmin penceresi sayısı kullanılarak pencere başına ortalama çıkarım süresi ve saniyede işlenen pencere sayısı hesaplanıp yazdırılır.

* **Uygulama Detayları/Sonuçlar (Beklenen):**
    * Toplam eğitim süresi yazdırılır.
    * Test seti üzerinde çıkarım yapılırken ilerleme çubuğu gösterilir.
    * Çıkarım tamamlandığında, toplam çıkarım süresi, tahmin penceresi başına ortalama çıkarım süresi ve saniyede işlenen tahmin penceresi sayısı ekrana yazdırılır. Bu, Autoformer modelinin tahmin hızını gösterir.

```python
# Hücre 11 (Autoformer): Eğitim ve Çıkarım Sürelerinin Hesaplanması

# time, torch, tqdm.auto, DataLoader zaten import edilmiş olmalı

# 1. Eğitim Süresi
if 'training_time_autoformer' in globals() and training_time_autoformer is not None:
    print(f"Toplam Autoformer Model Eğitim Süresi: {training_time_autoformer:.2f} saniye ({training_time_autoformer/60:.2f} dakika)")
else:
    print("Autoformer eğitim süresi bilgisi `training_time_autoformer` değişkeninden alınamadı.")
    print("Lütfen Hücre 7'nin (eğitim hücresi) başarıyla tamamlandığından emin olun.")

# 2. Tahmin Penceresi Başına Ortalama Çıkarım Süresi (Test Seti Üzerinden)
avg_inference_time_per_window_autoformer = None
if 'model_autoformer' in globals() and model_autoformer is not None and \
   'test_loader_af' in globals() and test_loader_af is not None and \
   'device' in globals() :

    model_autoformer.eval()

    total_inference_time_val_af = 0.0
    num_prediction_windows_processed = 0 # Artık "örnek" yerine "tahmin penceresi" diyoruz

    print(f"\nAutoformer Test seti üzerinde çıkarım süresi hesaplanıyor...")

    with torch.no_grad():
        for batch_data_test in tqdm(test_loader_af, desc="Autoformer Çıkarım"):
            # Veri yükleyiciden gelen 5 tensörü al (sonuncusu y_true, ilk 4'ü model girdisi)
            # batch_x_enc, batch_x_mark_enc, batch_x_dec, batch_x_mark_dec, _ = batch_data_test
            # ya da DataLoader'dan gelenleri doğru şekilde ayıralım
            test_x_enc, test_x_mark_enc, test_x_dec, test_x_mark_dec, _ = [d.to(device) for d in batch_data_test]


            start_batch_time = time.perf_counter()
            # Modelin forward metodunun sadece ilk 4 argümanı aldığını varsayıyoruz
            # (enc_self_mask, dec_self_mask, dec_enc_mask opsiyonel ve None)
            outputs = model_autoformer(test_x_enc, test_x_mark_enc, test_x_dec, test_x_mark_dec)
            if isinstance(outputs, tuple): outputs = outputs[0] # Ana tahmini al
            end_batch_time = time.perf_counter()

            batch_time = end_batch_time - start_batch_time
            total_inference_time_val_af += batch_time
            # Her bir batch, batch_size kadar tahmin penceresi içerir
            num_prediction_windows_processed += test_x_enc.size(0)

    if num_prediction_windows_processed > 0 and total_inference_time_val_af > 0 :
        avg_inference_time_per_window_autoformer = total_inference_time_val_af / num_prediction_windows_processed
        windows_per_second_inference_autoformer = num_prediction_windows_processed / total_inference_time_val_af
        print(f"\nToplam {num_prediction_windows_processed} tahmin penceresi için Autoformer çıkarım süresi: {total_inference_time_val_af:.4f} saniye")
        print(f"Tahmin penceresi başına ortalama Autoformer çıkarım süresi: {avg_inference_time_per_window_autoformer:.6f} saniye/pencere")
        print(f"Saniyede işlenen Autoformer tahmin penceresi sayısı (çıkarım): {windows_per_second_inference_autoformer:.2f} pencere/saniye")
    elif num_prediction_windows_processed > 0 and total_inference_time_val_af == 0 :
        print(f"\nToplam {num_prediction_windows_processed} tahmin penceresi için Autoformer çıkarım süresi çok kısa, hız çok yüksek.")
        avg_inference_time_per_window_autoformer = 0.0
    else:
        print("Autoformer Çıkarım için hiç tahmin penceresi işlenemedi veya süre ölçülemedi.")

else:
    print("Hata: Autoformer Modeli, test veri yükleyici veya cihaz bilgisi bulunamadı.")
```
